```python
def func(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'yandex':
        return 'organic'
    elif (row['traffic_source'] == 'paid' or row['traffic_source'] == 'paid') and row['region'] == 'Russia':
        return 'ad'
    elif (row['traffic_source'] == 'paid' or row['traffic_source'] == 'paid') and row['region'] != 'Russia':
        return 'other'
    return row['traffic_source']
```

```python
log['source_type'] = log.apply(func, axis=1)
```
